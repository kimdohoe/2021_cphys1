```python
import numpy as np
from matplotlib import pyplot as plt


m = 1.0
k = 1.0
C = 1.0
N = 40
w = 0.1


for i in range(0,4):

    w += 0.3
    d = (2*k - m*(w**2))

    M = []
    for r in range(0, N):
        row = []
        if r == 0:
            row.append(d-k)
            row.append(-k)
            for c in range(0, N-2):
                row.append(0)
        elif r == N-1:
            for c in range(0, N-2):
                row.append(0)
            row.append(-k)
            row.append(d-k)
        else:
            for c in range(0, r-1):
                row.append(0)
            row.append(-k)
            row.append(d)
            row.append(-k)
            for c in range(0, N-r-2):
                row.append(0)
        M.append(row)

    b = []
    b.append(C)
    for i in range(0, N-1):
        b.append(0)

    M_inv = np.linalg.inv(M)
    a = np.dot(M_inv, b)

    plt.title('w = 0.4(blue), 0.7(yellow), 1.1(green), 1.4(red)')
    plt.xlabel('i')
    plt.ylabel('a_i')
    plt.plot(a) 
```

# w 가 증가함에 따라서 i 대 a_i 그래프의 변동 주기가 짧아짐을 알 수 있다. #
# 예를 들어, w=0.4 일 때 변동 주기가 약 15 로 가장 길고, #
   # w=1.4일 때 변동 주기는 약 5로 가장 짧다 #
   # 이것은 주기적인 힘의 진동수와 i대 a_i 그래프의 진동수가 비례함을 의미한다 #
   # w 가 증가함에 따라서 그래프의 진폭은 감소하다가 w=1.4일 때 최대(약 12)를 형성한다 #
   # 이것은 힘의 진동수와 최대 가속도는 일반적으로는 반비례하지만  #
   # w=1.4 근방에서 공명진동이 일어나고 최대 가속도가 증가함을 의미한다 #

```python
import numpy as np
from matplotlib import pyplot as plt

m = 1.0
k = 1.0
C = 1.0
N = 40
w = 0.1


for i in range(0, 4):

    w += 0.3
    d = (2*k - m*(w**2))

    M = []
    for r in range(0, N):
        row = []
        if r == 0:
            row.append(d-k)
            row.append(-k)
            for c in range(0, N-2):
                row.append(0)
        elif r == N-1:
            for c in range(0, N-2):
                row.append(0)
            row.append(-k)
            row.append(d-k)
        else:
            for c in range(0, r-1):
                row.append(0)
            row.append(-k)
            row.append(d)
            row.append(-k)
            for c in range(0, N-r-2):
                row.append(0)
        M.append(row)

    b = []
    b.append(C)
    for i in range(0, N-2):
        b.append(0)
    b.append(C)

    M_inv = np.linalg.inv(M)
    a = np.dot(M_inv, b)

    plt.title('w = 0.4(blue), 0.7(yellow), 1.1(green), 1.4(red)')
    plt.xlabel('i')
    plt.ylabel('a_i')
    plt.plot(a)
```

# 힘이 양단에 가해지는 경우의 그래프는 진동수의 측면에서는 이전의 그래프와 유사하다. #
   # 차이점은 공명주파수가 아닐 때 최대 가속도가 이전의 그래프보다 현저히 낮고, #
   # 반대로 공명주파수일때의 최대 가속도(약 24)는 이전의 그래프(약 12)보다 현저히 높다 #
   # 이는 이전과 달리 양쪽에서 가해지는 힘에 의한 두 파동이  #
   # 비공명 진동에서는 상쇄간섭을 일으키고, 공명 진동에서는 보강간섭을 일으킴을 의미한다. #
   # 또한 공명진동에서의 최대가속도(약 24)가 이전(약 12)의 약 2배라는것은 #
   # 공명진동에서 상쇄없이(또는 매우 적게) 두 파동이 보강된다는 것을 의미한다 #
