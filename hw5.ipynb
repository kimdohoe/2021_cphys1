```python
import numpy as np
from matplotlib import pyplot as plt


m = 1.0
k = 1.0
C = 1.0
N = 40
w = 0.1


for i in range(0,4):

    w += 0.3
    d = (2*k - m*(w**2))

    M = []
    for r in range(0, N):
        row = []
        if r == 0:
            row.append(d-k)
            row.append(-k)
            for c in range(0, N-2):
                row.append(0)
        elif r == N-1:
            for c in range(0, N-2):
                row.append(0)
            row.append(-k)
            row.append(d-k)
        else:
            for c in range(0, r-1):
                row.append(0)
            row.append(-k)
            row.append(d)
            row.append(-k)
            for c in range(0, N-r-2):
                row.append(0)
        M.append(row)

    b = []
    b.append(C)
    for i in range(0, N-1):
        b.append(0)

    M_inv = np.linalg.inv(M)
    a = np.dot(M_inv, b)

    plt.title('w = 0.4(blue), 0.7(yellow), 1.1(green), 1.4(red)')
    plt.xlabel('i')
    plt.ylabel('a_i')
    plt.plot(a) 
```
