## 1.(a) ##

```python
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np
```

```python
def random_walk(n):
    elements = [1,-1]
    probabilities = [0.5,0.5]
    x = [0]
    for i in range(1, n+1):
        y = np.random.choice(elements, 1, p = probabilities)
        x.append(x[i-1]+y)    
    return x

n = 250
N = [i for i in range(n+1)]

plt.plot(N, random_walk(n))
plt.grid(axis = 'y')
plt.xlabel('n')
plt.ylabel('$X_n$')
plt.show()
```

## 1.(b) ##

```python
def mean(n):
    np.mean(random_walk(n))
    
    return np.mean(random_walk(n))
def var(n):
    np.var(random_walk(n))
    
    return np.var(random_walk(n))
def std(n):
    np.std(random_walk(n))
    
    return np.std(random_walk(n))
```

```python
np.mean(random_walk(n))
```

```python
np.var(random_walk(n))
```

```python
np.std(random_walk(n))
```

```python
rv = norm(loc = mean(n), scale = std(n))
x = np.arange(-20, 10, 0.01)
y = rv.pdf(x)
fig, ax = plt.subplots(1,1)
ax.plot(x,y,'bo',ms=1)
ax.vlines(x, 0, y, colors='black', lw = 5, alpha=0.5)
```
