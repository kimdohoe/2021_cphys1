```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import integrate
import math
```

```python
def y(x):
    return np.sin(x)

def trapez(f,a,b,n):
    s, x, dx = 0, a, (b-a)/n
    while(x<b):
        s += dx*(0.5*f(x)+0.5*f(x+dx))
        x = x + dx
    return s

trapez(y,0,np.pi,100)
```

```python
def y(x):
    return np.sin(x)
integrate.quad(y,0,np.pi*2)
```

```python
x = np.arange(0, 2*np.pi, 0.5)

plt.plot(x, y(x), color='black',marker='o')
plt.fill_between(x,y(x),alpha=0.2)
```

```python
z=[j for j in range(-20,20)]
c= [trapez(y,0,x,100) for x in range(-20,20)]
plt.plot(z,c)
```

```python
def g(x1):
    return np.cos(x1)

def trapez(f1,a1,b1,n1):
    s1, x1, dx1 = 0, a1, (b1-a1)/n1
    while(x1<b1):
        s1 += dx1*(0.5*f1(x1)+0.5*f1(x1+dx1))
        x1 = x1 + dx1
    return s1

trapez(y,0,np.pi*2,100)
```
